```python
def rmse(y_gold, y_pred): 
    return math.sqrt(((y_gold - y_pred)**2).mean())
```

```python
def print_score(m, X_train, y_train, X_valid, y_valid):
    print('RMSE on train set: {:.4f}'.format(rmse(m.predict(X_train), y_train)))
    print('RMSE on valid set: {:.4f}'.format(rmse(m.predict(X_valid), y_valid)))
    print('R^2 on train set: {:.4f}'.format(m.score(X_train, y_train)))
    print('R^2 on valid set: {:.4f}'.format(m.score(X_valid, y_valid)))
    if hasattr(m, 'oob_score_'): print('R^2 on oob set: {:.4f}'.format(m.oob_score_))
    return
```

## Xgboost MODEL

```python
def reg_perf(regressor, X_training, X_validation, y_true_training, y_true_validation, plot=True):
    """
    Evaluate the performance of a regression model.

    Parameters
    ----------
    regressor: sklearn.base.BaseEstimator. A trained scikit-learn regressor.
    X_training: pd.DataFrame. Table containing features used for training.
    X_validation: pd.DataFrame. Table containing features used for validation.
    y_true_training: array-like. List of true target variable values for training.
    y_true_validation: array-like. List of true target variable values for validation.
    plot: boolean. Whether to show a scatter plot of truth versus predictions.
    """
    y_pred_training = regressor.predict(X_training)
    y_pred_validation = regressor.predict(X_validation)
    rmse_training = metrics.root_mean_squared_error(y_true_training, y_pred_training)
    rmse_validation = metrics.root_mean_squared_error(y_true_validation, y_pred_validation)
    r2_training = metrics.r2_score(y_true_training, y_pred_training)
    r2_validation = metrics.r2_score(y_true_validation, y_pred_validation)
    print("RMSE on training set : {:.3f}".format(rmse_training))
    print("RMSE on validation set : {:.3f}".format(rmse_validation))
    print("R^2 on training set : {:.3f}".format(r2_training))
    print("R^2 on validation set : {:.3f}".format(r2_validation))

    if plot:
        sns.regplot(x=y_true_validation, y=y_pred_validation,  scatter_kws={'s': 1, 'alpha':0.7})
        plt.xlabel("true target variable")
        plt.ylabel("predicted target variable")
        plt.title("Regressor prediction vs truth on validation set")
        plt.show()
```

## CATBOOST MODEL

### Preprocessing for catboost

```python
df_cat = df.copy()
```

```python
nan_counts = df_cat.isnull().sum()
cols_to_drop = nan_counts[nan_counts > len(df_cat) * 0.4].index.tolist()
df_cat = df_cat.drop(columns=cols_to_drop)
```

```python
critical_cols = ['host_since', 'host_listings_count', 'host_total_listings_count', 
                 'host_has_profile_pic', 'host_identity_verified']
rows_to_drop = df_cat[critical_cols].isnull().any(axis=1)
df_cat = df_cat[~rows_to_drop]
```

```python
def clean_pct(x):
    if isinstance(x, str):
        return float(x.replace('%', '')) / 100
    return x
df_cat['host_response_rate'] = df_cat['host_response_rate'].apply(clean_pct)
df_cat['host_acceptance_rate'] = df_cat['host_acceptance_rate'].apply(clean_pct)
```

```python
cols_to_median = ['bathrooms', 'bedrooms', 'beds', 'host_response_rate', 'host_acceptance_rate']
for col in cols_to_median:
    if col in df_cat.columns:
        df_cat[f'{col}_filled_by_median'] = df_cat[col].isnull().astype(int)
        df_cat[col] = df_cat[col].fillna(df_cat[col].median())
```

```python
cols_to_unknown = ['has_availability', 'host_location', 'host_response_time', 'host_is_superhost']
for col in cols_to_unknown:
    if col in df_cat.columns:
        df_cat[col] = df_cat[col].fillna('Unknown')
```

```python
old_date = '2000-01-01'
df_cat['first_review'] = df_cat['first_review'].fillna(old_date)
df_cat['last_review'] = df_cat['last_review'].fillna(old_date)
```

```python
review_cols = [col for col in df_cat.columns if 'review_scores_' in col]
for col in review_cols:
    df_cat[col] = df_cat[col].fillna(0)

df_cat['reviews_per_month'] = df_cat['reviews_per_month'].fillna(0)
```

```python
for col in ['first_review', 'last_review', 'host_since']:
    df_cat[col] = pd.to_datetime(df_cat[col])
    df_cat[col] = df_cat[col].apply(lambda x: x.toordinal() if pd.notnull(x) else 0)
```

```python
def group_verifications(x):
    x_str = str(x)
    if x_str == "['email', 'phone']":
        return 'email_phone'
    elif x_str == "['email', 'phone', 'work_email']":
        return 'email_phone_work'
    elif x_str == "['phone', 'work_email']":
        return 'phone_work'
    elif x_str == "['phone']":
        return 'phone_only'
    elif x_str == "['email']":
        return 'email_only'
    elif x_str == "[]" or x_str == "nan" or pd.isna(x):
        return 'none'
    else:
        return 'other'

df_cat['verification_group'] = df_cat['host_verifications'].apply(group_verifications)
df_cat = df_cat.drop(columns=['host_verifications'])
```

```python
cols_to_keep = [
    'host_since', 'host_response_rate', 'host_acceptance_rate',
    'host_listings_count', 'host_total_listings_count',
    'latitude', 'longitude', 'accommodates', 'bathrooms', 'bedrooms', 'beds',
    'availability_30', 'availability_60', 'availability_90', 'availability_365',
    'number_of_reviews', 'number_of_reviews_ltm', 'number_of_reviews_l30d',
    'first_review', 'last_review', 'reviews_per_month',
    
    'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
    'review_scores_checkin', 'review_scores_communication', 'review_scores_location',
    'review_scores_value',
    
    'host_response_time', 'host_is_superhost', 'host_has_profile_pic',
    'host_identity_verified', 'room_type', 'has_availability', 'city',
    'property_type', 'verification_group',
    
    'bathrooms_filled_by_median', 'bedrooms_filled_by_median', 'beds_filled_by_median',
    'host_response_rate_filled_by_median', 'host_acceptance_rate_filled_by_median',
    
    'price'
]

cols_to_keep = [col for col in cols_to_keep if col in df_cat.columns]
df_cat = df_cat[cols_to_keep]
```

```python
categorical_features = [
    'host_response_time', 'host_is_superhost', 'host_has_profile_pic',
    'host_identity_verified', 'room_type', 'has_availability', 'city',
    'property_type', 'verification_group'
]
categorical_features = [col for col in categorical_features if col in df_cat.columns]

X_cat = df_cat.drop(columns=['price'])
y_cat = df_cat['price']
y_cat_log = np.log1p(y_cat)

X_train_cat, X_valid_cat, y_train_cat_log, y_valid_cat_log = train_test_split(
    X_cat, y_cat_log, test_size=0.2, random_state=42
)
```

```python
for col in categorical_features:
    X_train_cat[col] = X_train_cat[col].astype(str)
    X_valid_cat[col] = X_valid_cat[col].astype(str)

```

### Model

```python
def objective_cat(trial):
    catboost_model = CatBoostRegressor(
        iterations=trial.suggest_int('iterations', 2000, 4000),
        depth=trial.suggest_int('depth', 5, 8),
        learning_rate=trial.suggest_float('learning_rate', 0.01, 0.03),
        l2_leaf_reg=trial.suggest_int('l2_leaf_reg', 5, 12),
        bagging_temperature=trial.suggest_float('bagging_temperature', 0.5, 1.0),
        subsample=trial.suggest_float('subsample', 0.8, 0.9),
        random_strength=trial.suggest_int('random_strength', 1, 5),
        loss_function='RMSE',
        random_seed=42,
        verbose=0
    )
    catboost_model.fit(
        X_train_cat, y_train_cat_log,
        eval_set=(X_valid_cat, y_valid_cat_log),
        cat_features=categorical_features,
        early_stopping_rounds=120
    )
    pred = catboost_model.predict(X_valid_cat)
    rmse = np.sqrt(mean_squared_error(y_valid_cat_log, pred))
    r2 = r2_score(y_valid_cat_log, pred)
    print(f"CatBoost Trial {trial.number}: RMSE={rmse:.4f}, R²={r2:.4f}")
    return rmse
```

```python
study_cat = optuna.create_study(direction='minimize')
study_cat.optimize(objective_cat, n_trials=20)
print("Best CatBoost params:", study_cat.best_params)
```

```python
best_params = study_cat.best_params

catboost_model = CatBoostRegressor(
    iterations=best_params['iterations'],
    depth=best_params['depth'],
    learning_rate=best_params['learning_rate'],
    l2_leaf_reg=best_params['l2_leaf_reg'],
    bagging_temperature=best_params['bagging_temperature'],
    subsample=best_params['subsample'],
    random_strength=best_params['random_strength'],
    loss_function='RMSE',
    random_seed=42,
    verbose=100
)

catboost_model.fit(
    X_train_cat, y_train_cat_log,
    eval_set=(X_valid_cat, y_valid_cat_log),
    cat_features=categorical_features,
    early_stopping_rounds=120
)
```

### Score

```python

y_pred_train = catboost_model.predict(X_train_cat)
y_pred_valid = catboost_model.predict(X_valid_cat)

rmse_train = math.sqrt(mean_squared_error(y_train_cat_log, y_pred_train))
rmse_valid = math.sqrt(mean_squared_error(y_valid_cat_log, y_pred_valid))
r2_train = r2_score(y_train_cat_log, y_pred_train)
r2_valid = r2_score(y_valid_cat_log, y_pred_valid)

print(f"RMSE (train): {rmse_train:.4f}")
print(f"RMSE (valid): {rmse_valid:.4f}")
print(f"R² (train): {r2_train:.4f}")
print(f"R² (valid): {r2_valid:.4f}")
```
